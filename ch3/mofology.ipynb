```python
import cv2 as cv
import sys
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv.imread('hockey.jpeg')
if img is None:
    sys.exit('Could not read the image.')
```

```python
t,bin_img=cv.threshold(img[:,:,0],0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)
plt.imshow(bin_img,cmap='gray'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
b=bin_img[0:bin_img.shape[0]//2+1,bin_img.shape[1]//2:bin_img.shape[1]]
plt.imshow(b,cmap='gray'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
se=np.uint8([[0,1,0],[1,1,1],[0,1,0]])
```

```python
dilated=cv.dilate(b,se,iterations=1)
plt.imshow(dilated,cmap='gray'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
eroded=cv.erode(dilated,se,iterations=1)
plt.imshow(eroded,cmap='gray'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
closing=cv.erode(dilated,se,iterations=1)
plt.imshow(closing,cmap='gray'), plt.xticks([]), plt.yticks([])
plt.show()
```
