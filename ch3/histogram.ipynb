```python
import cv2 as cv
import sys
import matplotlib.pyplot as plt
```

```python
img=cv.imread('hockey.jpeg')
if img is None:
    sys.exit('Could not read the image.')
```

```python
gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='gray'), plt.xticks([]), plt.yticks([]),plt.show()
```

```python
h=cv.calcHist([img],[1],None,[256],[0,256])
plt.plot(h,color='r',linewidth=2)
```

```python
equalized=cv.equalizeHist(gray)
plt.imshow(equalized,cmap='gray'), plt.xticks([]), plt.yticks([]),plt.show()
```

```python
h=cv.calcHist([equalized],[0],None,[256],[0,256])
plt.plot(h,color='r',linewidth=2), plt.show()
```
